```python
import pandas as pd
import numpy as np

# Daten für die Übungsaufgaben
daten = {
    'Name': ['Anna', 'Mark', 'Julia', 'Tom', 'Laura', 'Peter', 'Sofia'],
    'Abteilung': ['IT', 'Sales', 'IT', 'HR', 'Sales', 'IT', 'HR'],
    'Gehalt': [70000, 62000, 75000, 50000, 68000, 69000, 52000], # Gehalt für Peter gefüllt
    'HomeOfficeTage': [3, 2, 4, 1, 3, 3, 2]
}
index_ids = ['M-101', 'M-102', 'M-103', 'M-104', 'M-105', 'M-106', 'M-107']
df_mitarbeiter = pd.DataFrame(daten, index=index_ids)

print("Sauberer Mitarbeiter-DataFrame für die Übungen:")
print(df_mitarbeiter)
```

Aufgabe 5.1: Neue Spalte berechnen

Erstelle eine neue Spalte namens Jahresgehalt, indem du die Spalte Gehalt mit 12 multiplizierst.


```python
df_mitarbeiter['Jahresgehalt'] = df_mitarbeiter['Gehalt'] * 12
df_mitarbeiter
```


___
Aufgabe 5.2: Spalte umbenennen

Der Spaltenname HomeOfficeTage ist etwas lang. Benenne die Spalte in HO_Tage um. Gib den DataFrame danach aus, um die Änderung zu sehen.


```python
df_mitarbeiter.rename({'HomeOfficeTage': 'HO_Tage'}, inplace=True, axis='columns')
df_mitarbeiter
```


___
Aufgabe 5.3: Spalte entfernen

Erstelle einen neuen DataFrame df_gefiltert, der eine Kopie von df_mitarbeiter ist, aber ohne die Spalte HO_Tage.


```python
df_gefiltert = df_mitarbeiter.drop('HO_Tage', axis=1).copy()
df_gefiltert
```


___
Aufgabe 5.4: Eine Bewertungsfunktion mit .apply()

Schreibe eine kleine Python-Funktion bewerte_gehalt(gehalt). Die Funktion soll 'Hoch' zurückgeben, wenn das Gehalt über 65000 liegt, und ansonsten 'Standard'.
Wende diese Funktion mithilfe von .apply() auf die Gehalt-Spalte an, um eine neue Spalte namens Gehaltsklasse zu erstellen.
(Diese Aufgabe dient als Ausgangspunkt für die effizienteren Alternativen im nächsten Teil.)


```python
def bewerten(gehalt: int) -> bool:
    if gehalt > 65000:
        return True
    else:
        return False
    
df_mitarbeiter['Gehaltsklasse'] = df_mitarbeiter['Gehalt'].apply(bewerten)
df_mitarbeiter
```


___
Teil 2: Effiziente Alternativen zu .apply() (apply_alternatives.ipynb)
Aufgabe 5.5: Effiziente Alternative mit np.where()

Erstelle die Spalte Gehaltsklasse aus Aufgabe 5.4 erneut, aber verwende dieses Mal die deutlich schnellere Funktion np.where() anstelle von .apply().


```python
import numpy as np

df_mitarbeiter['Gehaltsklasse2'] = np.where(df_mitarbeiter['Gehalt'] > 65000, True, False)
df_mitarbeiter
```


___
Aufgabe 5.6: Werte ersetzen mit .map()

Erstelle ein Python-Dictionary abteilungs_codes, das die Abteilungen auf Kürzel abbildet: {'IT': 'TECH', 'Sales': 'SALE', 'HR': 'PERS'}.
Verwende die .map()-Methode, um basierend auf der Abteilung-Spalte eine neue Spalte Abteilungs_Code zu erstellen, die diese Kürzel enthält.


```python
abt = {'IT': 'TECH', 'Sales': 'SALE', 'HR': 'PERS'}

df_mitarbeiter['Abteilungs_code'] = df_mitarbeiter['Abteilung'].map(abt)
df_mitarbeiter
```


___
Aufgabe 5.7: String-Operationen mit dem .str-Accessor

Erstelle eine neue Spalte Name_Gross, die die Namen aller Mitarbeiter in Großbuchstaben enthält. Verwende hierfür den vektorisierten .str-Accessor.


```python
df_mitarbeiter['Name_Gross'] = df_mitarbeiter['Name'].str.upper()
df_mitarbeiter['Name_Lower'] = df_mitarbeiter['Name'].str.lower()   
df_mitarbeiter['Name_Titel'] = df_mitarbeiter['Name'].str.title()
df_mitarbeiter['Name_Laenge'] = df_mitarbeiter['Name'].str.len()
df_mitarbeiter['Name_Anzahl_A'] = df_mitarbeiter['Name'].str.count('a')
df_mitarbeiter
```


___
Aufgabe 5.8 (Bonus-Challenge): Kategorisierung mit pd.cut()

Die Personalabteilung möchte die Mitarbeiter basierend auf ihrem Gehalt in drei gleich große Gruppen einteilen: 'Gering', 'Mittel', 'Top'.
Verwende pd.cut(), um eine neue Spalte Gehaltsgruppe zu erstellen, die diese drei Kategorien enthält.
(Tipp: pd.cut() kann anstelle einer Liste von Grenzen auch einfach eine Ganzzahl für die gewünschte Anzahl an "Bins" entgegennehmen.)

```python
df_mitarbeiter['Gehaltsgruppe'] = pd.cut(df_mitarbeiter['Gehalt'], 3, labels=['Gering', 'Mittel', 'Hoch'])
df_mitarbeiter
```
